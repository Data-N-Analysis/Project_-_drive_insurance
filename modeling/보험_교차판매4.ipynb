```python
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold, RandomizedSearchCV, train_test_split
from sklearn.metrics import accuracy_score, classification_report
import xgboost as xgb
import matplotlib.pyplot as plt
from scipy.stats import uniform, randint
```

```python
# 3. XGBoost 모델 정의 (n_estimators 고정)
model = xgb.XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='mlogloss')

# 4. 하이퍼파라미터 분포 설정 (n_estimators 제외)
param_dist = {
    'max_depth': randint(3, 7),
    'learning_rate': uniform(0.01, 0.2),
    'subsample': uniform(0.7, 0.9)
}

# 5. K-Fold 교차 검증 설정
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 6. RandomizedSearchCV를 사용한 하이퍼파라미터 튜닝
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=15, cv=kf, scoring='accuracy', n_jobs=-1, verbose=1, random_state=42)
random_search.fit(X_train, Y_train)

# 7. 최적의 하이퍼파라미터 출력
print(f"최적의 하이퍼파라미터: {random_search.best_params_}")
```

```python
# 8. 최적 모델로 예측 및 성능 평가
best_model = random_search.best_estimator_
Y_pred = best_model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(f"테스트 세트 정확도: {accuracy:.4f}")
print(f"분류 보고서:\n{classification_report(Y_test, Y_pred)}")

# 9. 피처 중요도 시각화
importance = best_model.feature_importances_
features = data.feature_names

plt.figure(figsize=(10, 6))
plt.barh(features, importance, color='b')
plt.xlabel('Feature Importance')
plt.title('Feature Importance in XGBoost Model')
plt.show()
```
